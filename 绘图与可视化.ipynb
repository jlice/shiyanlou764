```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib as mpl
from pandas import Series, DataFrame

%matplotlib inline
```

```python
# 创建Figure对象
fig = plt.figure()

# 创建轴域
ax1 = fig.add_subplot(2,2,1)  # 2,2,1 表示在两行两列的第1个位置上
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)

# 绘图
ax1.plot(np.random.randn(50).cumsum(), 'g--')
ax2.hist(np.random.randn(100), bins=20, color='m')
ax3.scatter(np.arange(30), np.arange(30)+3*np.random.randn(30))
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

ax.set_title('Axes Example')

# ticks 是坐标轴上的刻度
major_ticks = np.arange(0, 101, 20)
minor_ticks = np.arange(0, 101, 5)

ax.set_xticks(major_ticks)
ax.set_xticks(minor_ticks, minor=True)
ax.set_yticks(major_ticks)
ax.set_yticks(minor_ticks, minor=True)

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')

ax.grid(which='minor', alpha=0.2)
ax.grid(which='major', alpha=0.5)

ax.text(42, 40 , 'shiyanlou')
```

```python
x = np.linspace(0, 1, 100)

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title('shiyanlou')

# label可以使用LaTeX数学公式 
ax.plot(x, x**(1/8), 'b--', label=r'$y=x^{\frac {1} {8}}$')
ax.plot(x, x**8, 'c*', label=r'$y=x^{8}$')
ax.plot(x, x**(1/2), 'g', label=r'$y=x^{\frac {1} {2}}$')
ax.plot(x, x**2, 'r', label=r'$y=x^{2}$')
ax.plot(x, x, 'k', label=r'$y=x$')

ax.axis([0,1,0,1])  # 设置坐标轴范围，[xmin, xmax, ymin, ymax]
ax.legend()  # 创建图例
```

```python
# 箱线图可以看出数据的分散程度、异常值等信息
spread = np.random.rand(50)*100
center = np.ones(25)*50
outlier_high = np.random.rand(10)*100+100
outlier_low = np.random.rand(10)*(-100)
data = np.concatenate((spread, center, outlier_high, outlier_low), 0)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.boxplot(data)
fig.show()
```

```python
# pandas高度整合matplotlib，可以直接在Series上绘制线形图
s = Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
# 可以直接调用DataFrame的方法进行绘图
df = pd.DataFrame(np.random.rand(5,4), columns=list('ABCD'))
df.boxplot()
```

```python
# 使用Seaborn进行绘图
sns.set()

x = np.linspace(0, 14, 100)
for i in range(1, 7):
    plt.plot(x, np.sin(x+i*.5)*(7-i)*1)
```

```python
# 使用Bokeh绘图
from bokeh.layouts import gridplot
from bokeh.plotting import figure, show, output_file

x = np.linspace(0, 4*np.pi, 1000)
y = np.sin(x)

p = figure(title='Shiyanlou Bokeh Example')

p.circle(x, y, legend='sin(x)')
p.circle(x, 2*y, legend='2*sin(x)', color='red')
p.circle(x, 3*y, legend='3*sin(x)', color='green')

show(p)

# 不是基于matplotlib， 而是基于D3.js，在浏览器中显示
# 运行后会生成网页
```
